```python
import pandas as pd
data = pd.read_csv('ecg.csv', header=None)
# print(data)
# raw = data.values
data.head().style.set_properties(**{'background-color': 'black', 'color': 'white', 'border-color': 'white'})
```

```python
from sklearn.model_selection import train_test_split
import numpy as np
import tensorflow as tf

data = np.loadtxt("ecg.csv", delimiter=",")

features = data[:, :-1]  
labels = data[:, -1]    

x_train, x_test, y_train, y_test = train_test_split(features, labels, test_size=0.2)

model = tf.keras.models.Sequential([
  tf.keras.layers.Reshape((140, 1), input_shape=(140,1)),
  tf.keras.layers.Conv1D(filters=32, kernel_size=3, activation="relu"),
  tf.keras.layers.MaxPooling1D(pool_size=2),
  tf.keras.layers.Flatten(),
  tf.keras.layers.Dense(units=1, activation="sigmoid")
])

model.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])
model.fit(x=x_train, y=y_train, epochs=100)

model.evaluate(x=x_test, y=y_test)
model.save("model_train.h5")
```

```python
# Load m hnh  c hun luyn
model = tf.keras.models.load_model("model_train.h5")

# c d liu t file single_ecg_input.csv
single_ecg_test = pd.read_csv("normal_ecg.csv", header=None)

# Chuyn i nh dng d liu thnh mng numpy
single_ecg_test = single_ecg_test.values.reshape((1, 140))

# Repredict
prediction = model.predict(single_ecg_test)

# In ra d on
print(prediction)

# Kim tra v in ra kt qu d on
if prediction[0] > 0.5:
    print("UNKNOWN")
else:
    print("NORMAL ECG")
```
